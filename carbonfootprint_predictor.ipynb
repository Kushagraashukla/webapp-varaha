```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MultiLabelBinarizer

mlb = MultiLabelBinarizer()

main_data['Recycling'] = main_data['Recycling'].apply(lambda x: x.split(',') if isinstance(x, str) else [])
main_data['Cooking_With'] = main_data['Cooking_With'].apply(lambda x: x.split(',') if isinstance(x, str) else [])

recycling_encoded = pd.DataFrame(mlb.fit_transform(main_data['Recycling']), columns=['Recycling_' + c for c in mlb.classes_])
cooking_encoded = pd.DataFrame(mlb.fit_transform(main_data['Cooking_With']), columns=['CookingWith_' + c for c in mlb.classes_])

main_data = pd.concat([main_data, recycling_encoded, cooking_encoded], axis=1)

main_data.head()
```

```python
le=LabelEncoder()
```

```python
main_data['Sex']=le.fit_transform(main_data['Sex'])
```

```python
#one hot encoding
main_data_encoded = pd.get_dummies(main_data, columns=['Body Type', 'Diet', 'Heating Energy Source','How Often Shower','Transport','Vehicle Type','Social Activity','Waste Bag Size','Frequency of Traveling by Air','Energy efficiency'], drop_first=True)
main_data_encoded.head()
```

```python
main_data_encoded = main_data_encoded.drop(['Recycling','Cooking_With'], axis=1)
```

```python
main_data_encoded.info()
```

**FEATURE SELECTION**

```python
#since the features like internet use and tv laptop have least impact on carbon emission we can drop them
main_data_encoded = main_data_encoded.drop(['How Long TV PC Daily Hour', 'How Long Internet Daily Hour'], axis=1)
```

**SPLITING THE DATASET INTO FEATURES AND TARGET VARIABLE**

```python
Y = main_data_encoded['CarbonEmission']
X = main_data_encoded.drop('CarbonEmission', axis=1)
```

**splitting the data into training and test data**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
# Basic regressors
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Advanced regressors
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
```

```python
models=[LinearRegression(),Ridge(),Lasso(),DecisionTreeRegressor(),RandomForestRegressor(),XGBRegressor(),LGBMRegressor(),CatBoostRegressor(verbose=0)]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
def compare_model_cross_validation():
    # Convert DataFrame to numpy array
    X_array = X.to_numpy()
    Y_array = Y.to_numpy()
    
    print("=== Cross Validation Scores ===")
    for model in models:
        cv_score = cross_val_score(model, X_array, Y_array, cv=5)
        mean_accuracy = cv_score.mean() * 100
        model_name = model.__class__.__name__
        print(f"{model_name}: Scores = {cv_score}, Mean Accuracy = {mean_accuracy:.2f}%")
```

```python
compare_model_cross_validation()
```

**so from here we can conclude that the CatBoostRegressor is the best model with accuracy of 98.89%**

Training the model

```python
cat=CatBoostRegressor()
```

```python
cat.fit(X_train,Y_train)
# Train ke baad
cat.save_model("catboost_model.cbm")
```
